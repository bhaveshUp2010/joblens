## We work Remotely

```python
headers = {
    "User-Agent": "Mozilla/5.0"
}
base_url = "https://weworkremotely.com"
url = f"{base_url}/categories/remote-programming-jobs"
res = requests.get(url, headers = headers)

soup = BeautifulSoup(res.text, "xml")
print(soup.title.text)
```

```python
jobs = soup.find_all("item")
for job in jobs:
    title = job.title.text if job.title else "N/A"
    region = job.region.text if job.region else "N/A"
    category = job.category.text if job.category else "N/A"
    description = job.description.text if job.description else "N/A"
    link = job.link.text if job.link else "N/A"
    print("Title:", title)
    print("Category:", category)
    print("Region:", region)
    print("Link:", link)
    print("Description preview:", description[:200])
    print("-" * 60)
```

## Hack News

```python
base_url = "https://news.ycombinator.com/item?id="
post_id = "46857488"
url = base_url+post_id

headers = {
    "User-Agent": "Mozilla/5.0"
}
res = requests.get(url, headers = headers)

soup = BeautifulSoup(res.text, "html.parser")
```

```python
comments = soup.find_all("tr", class_="athing comtr")

for c in comments:
    user_tag = soup.find("a", class_="hnuser")
    user = user_tag.text if user_tag else "N/A"

    text_tag = soup.find("div", class_="commtext")
    if not text_tag:
        continue

        raw_text = text_tag.get_text("\n", strip = True)
    print({user, raw_text})
```

## Muse

```python
url = "https://www.themuse.com/api/public/jobs"

headers = {
    'Content-Type': 'application/json'
}
params = {
    "page": 1
}

res = requests.get(url, headers=headers, params= params)

data = res.json()

jobs = data.get("results", [])

for job in jobs:
    title = job.get("name")
    company = job.get("company").get("name")
    location = job.get("locations")

    html = job.get("contents")
    soup = BeautifulSoup(html, "html.parser")
    description = soup.get_text(separator= " ")
    print({
        "title": title,
        "company": company,
        "location": location,
        "desc": description[:200]
    })
    break
```
